### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# YOUR CHANGES HERE

while len(q10) < 10:
    x = q10[-1]
    if x % 2 == 0:
        q10.append(x // 2)
    else:
        q10.append(3 * x + 1)


plt.figure(figsize=(10, 6))

plt.plot(q10, marker="o", linestyle="-", color="blue", label="Collatz sequence")
plt.xlabel("Step")
plt.ylabel("Value")
plt.title("Collatz sequence starting from 15")
plt.legend()
plt.grid(True)

plt.xticks(range(len(q10)), [str(i + 1) for i in range(len(q10))])
plt.show()
````

### Problem 15

Use the function `math.comb` to make a bar chart illustrating the binomial distribution for $n=10$.
That is, for integers x from zero to ten, there should be a bar whose height is $10 \choose x$.

````python
# YOUR CHANGES HERE

import math
import matplotlib.pyplot as plt


n = 10

x_values = list(range(n + 1))
y_values = [math.factorial(n) // (math.factorial(x) * math.factorial(n - x)) for x in x_values]


plt.figure(figsize=(10, 6))
plt.bar(x_values, y_values, color='skyblue', edgecolor='black')
plt.xlabel('$x$ (Number of Successes)', fontsize=14)
plt.ylabel(r'Binomial Coefficient $\binom{10}{x}$', fontsize=14)
plt.title(r'Binomial Distribution Coefficients for $n=10$: $\binom{10}{x}$', fontsize=16)
plt.xticks(x_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

### Problem 17

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# YOUR CHANGES HERE
import matplotlib.pyplot as plt
import math

x = list(range(1, 11))
y = [math.factorial(n) for n in x]


plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.yscale("log")  
plt.xlabel("n")
plt.ylabel("n!")
plt.title("Factorial Function (Log Scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.7)
plt.show()
````
